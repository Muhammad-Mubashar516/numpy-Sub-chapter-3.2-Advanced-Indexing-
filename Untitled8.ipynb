```python
import numpy as np

print("\n----- Sub-chapter 3.2: Advanced Indexing -----")

# -------------------------------------------------------
# Topic 1: Integer Array Indexing
# -------------------------------------------------------
arr = np.array([10, 20, 30, 40, 50])
print("\nTopic 1: Integer Array Indexing:")
indices = [0, 2, 4]
print("Selected elements:", arr[indices])   # 10, 30, 50

# -------------------------------------------------------
# Topic 2: Boolean Indexing
# -------------------------------------------------------
print("\nTopic 2: Boolean Indexing:")
mask = arr > 25
print("Mask:", mask)
print("Filtered values:", arr[mask])   # values > 25

# -------------------------------------------------------
# Topic 3: Fancy Indexing
# -------------------------------------------------------
arr2 = np.array([[10, 20, 30],
                 [40, 50, 60],
                 [70, 80, 90]])

print("\nTopic 3: Fancy Indexing:")
rows = [0, 2]
cols = [1, 2]
print("arr2[rows, cols] =", arr2[rows, cols])
# â†’ (0,1) = 20  and (2,2) = 90

# -------------------------------------------------------
# Topic 4: Combining Indexing Types
# -------------------------------------------------------
print("\nTopic 4: Combining Indexing Types:")
print(arr2[0:2, [0, 2]])
# first 2 rows & columns [0,2]

# -------------------------------------------------------
# Topic 5: Indexing with Lists
# -------------------------------------------------------
print("\nTopic 5: Indexing with Lists:")
lst_idx = [1, 1, 2, 0]
print("arr[lst_idx] =", arr[lst_idx])

# -------------------------------------------------------
# Topic 6: Indexing with Arrays
# -------------------------------------------------------
print("\nTopic 6: Indexing with Arrays:")
idx_arr = np.array([2, 0, 3])
print("arr[idx_arr] =", arr[idx_arr])

# -------------------------------------------------------
# Topic 7: np.take()
# -------------------------------------------------------
print("\nTopic 7: np.take():")
print("np.take(arr, [1,3,4]) =", np.take(arr, [1, 3, 4]))

# -------------------------------------------------------
# Topic 8: np.put()
# -------------------------------------------------------
print("\nTopic 8: np.put():")
arr_copy = arr.copy()
np.put(arr_copy, [1, 3], [999, 555])
print("After np.put:", arr_copy)

# -------------------------------------------------------
# Topic 9: Masked Arrays (np.ma)
# -------------------------------------------------------
print("\nTopic 9: Masked Arrays (np.ma):")
data = np.array([5, -1, 8, -3, 10])
masked = np.ma.masked_less(data, 0)   # mask all negative values
print("Masked array:", masked)
print("Data without masked values:", masked.compressed())  # remove masked
```
